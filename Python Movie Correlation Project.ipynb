```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read In The Data 

df = pd.read_csv(r'C:\Users\jacks\OneDrive\Documents\data analyst portfolio proj\Correlation Project In Python\archive\movies.csv')
```

```python
# Let's look at the data!

df.head()
```

```python
# Dropping rows with no data

df = df.dropna()
```

```python
df.head()
```

```python
# Checking if there is missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Changing the Data types in our columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#Create the correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# Make df show in ascending order the highest gross revenue films

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# Drop any duplicates in the data

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Predicting Correlations:
#Budget will correlate with high revenue
#Company will correlate to high revenue
```

```python
# Scatter Plot w/ budget vs. gross revenue!

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')

plt.show()
```
